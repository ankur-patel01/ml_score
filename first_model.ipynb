<h4 style="color:green"> <i>Importing libraries</i></h4>

```python
#importing all libraries required for this notebook
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt
from matplotlib import style
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

<h4 style="color:green"> <i>Reading Data</i></h4>

```python
#reading a csv file
data_set=pd.read_csv("marks.csv")
print("The data imported, as shown below")
print(data_set.head())
```

<h4 style="color:green"> <i>Plotting Dataset to see and find any relationship between the Data</i></h4>

```python
#plotting the distribution of scores column in dataset
style.use('ggplot')
data_set.plot(x='Hours',y='Scores',style='o')
plt.title("HOURS VS SCORES")
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.legend()
plt.show()
```

### as we can see from the above graph that the relation in data is linear in positive direction

<h4 style="color:green"> <i>Preparing the data for prediction</i></h4>

```python
#dividing data into inputs and Outputs 
#that is features and target
X= data_set.iloc[:, :-1].values  
y= data_set.iloc[:, 1].values
```

<h4 style="color:green"> <i>Splitting the data into Training and Testing in the ratio 3 to 1 </i></h4>

```python
#splitting into X_train,X_test,y_train,y_test
X_train,X_test, y_train,y_test=train_test_split(X,y,test_size=1/3,random_state=1)
```

<h3 style="color:green"> <i>Building Model</i></h3>
<h4 style="color:green"> <i>Defining and Fitting Model</i></h4>


```python
reg=LinearRegression()
reg.fit(X_train,y_train)
print("Training of Model has been completed.")
```

```python
#plotting the line
line=reg.coef_*X+reg.intercept_
plt.scatter(X,y,color='red',label='Actual')
plt.plot(X,line,color='green', label='training')
plt.title("Hours Vs Score")
plt.xlabel('Study Hours')
plt.ylabel('Marks(Out of 100)')
plt.legend()
plt.show()
```

<h4 style="color:green"> <i>Making Predictions on Testing Data</i></h4>

```python
#making predictions using predict()
pred1=reg.predict(X_test)
print(pred1)
```

<h4 style="color:green"> <i>Comparing the actual and Predicted Data  </i></h4>

```python
#plotting the points to compare Testing Data
plt.scatter(X_test,y_test,color='red' ,label='Actual')
plt.plot(X_test,pred1,color='green',label='Predicted')
plt.title("Hours Vs Score (Testing Data)")
plt.xlabel('Study Hours')
plt.ylabel('Marks(Out of 100)')
plt.legend()
plt.show()
```

<h4 style="color:green"> <i>Predicting the custom data</i></h4>

## What will be predicted score if a student studies for 9.25 hrs/ day?

```python
hours_per_day=[[9.25]]
pred_cust=reg.predict(hours_per_day)
print("His Predicted Score is =")
print(pred_cust)
```

<h4 style="color:green"> <i>Validating the Model</i></h4>

```python
#Evaluating the Model accuracy
precision=mean_absolute_error(y_test,pred1)
print("Mean Absolute Error of this Model =")
print(precision)
```
